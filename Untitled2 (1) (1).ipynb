```python
#2.4
def transformation_plot(pts,matrix):
    figure,graph=pyplot.subplots(1,2,figsize=(10,5))
    x=pts[:,0]
    y=pts[:,1]
    c1 = x + y
    graph[0].scatter(x,y,100,c=c1)
    d=pts.dot(matrix)
    x=d[:,0]
    y=d[:,1]
    graph[1].scatter(x,y,100,c=c1)
transformation_plot(numpy.random.random(size=(100, 2)),numpy.array([[6,5],[4,1]]))
```

```python
#2.5
figure,graph=pyplot.subplots(1,2,figsize=(10,5))
x=numpy.linspace(0,5,100)
y=[numpy.sin(i) for i in x]
x1=x[0:-1]-x[1:]
y1=numpy.array(y[0:-1])-numpy.array(y[1:])
y11=numpy.divide(y1, x1)
x11=(x[0:-1]+x[1:])/2
graph[0].scatter(x11,y11,100)
graph[0].scatter(x,numpy.cos(x),100)
```

```python
#2.6
v=numpy.random.random(size=(100,2))-0.5
q=numpy.array([2,1])
counter90=0
counter30=0
for row in v:
    cos=row.dot(q)/(numpy.linalg.norm(q)*numpy.linalg.norm(row))
    if numpy.arccos(cos)*180/numpy.pi<90:
        counter90=counter90+1
    if numpy.arccos(cos)*180/numpy.pi<30:
        counter30=counter30+1
print(' ,   (v,q) < 90 :', counter90/100, '\n ,   (v,q) < 30 :', counter30/100)
```

```python
#2.7
res=[]
for i in range(2,15):
    a=0
    pts=numpy.random.random(size=(1000,i))
    for j in pts:
        if 1 >= sum([k**2 for k in j]):
            a=a+1
    res.append(a/1000)
figure,graph=pyplot.subplots(figsize=(10,5))
graph.set_xlim(2,15)
graph.set_ylim(0,0.85)
graph.plot(range(2,15),res,100)
```
